```python
df.drop(['body'],axis=1, inplace=True)
df.shape
```

```python
# simple impution
from sklearn.impute import SimpleImputer
```

```python
print(f'number of null before imputing :{df.age.isnull().sum()}')
```

```python
imp=SimpleImputer(strategy='mean')
df.age=imp.fit_transform(df[['age']])
print(df.age.isnull().sum())
```

```python
def get_parameters(df):
    parameters = {}
    for col in df.columns[df.isnull().any()]:  # Only columns with NaN
        if df[col].dtype in ['float64', 'int64', 'int32']:
            strategy = 'mean'  # Use mean for numeric columns
        else:
            strategy = 'most_frequent'  # Use mode for categorical columns

        # Always get the first missing value
        missing_values = df[col][df[col].isnull()].values[0]

        parameters[col] = {
            'missing_values': missing_values,
            'strategy': strategy
        }

    return parameters
get_parameters(df)
```

```python
parameters = get_parameters(df)

for col, param in parameters.items():
    missing_values = param['missing_values']
    strategy = param['strategy']

    imp = SimpleImputer(missing_values=missing_values, strategy=strategy)
    df[col] = imp.fit_transform(df[[col]]).ravel()  # fit_transform used here!
```
